```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import roc_curve, auc

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data (1000 samples)
n_samples = 1000
y_true = np.random.randint(0, 2, n_samples)  # True labels (50% class 0, 50% class 1)
y_scores = np.zeros(n_samples)

# Create realistic score distributions
# Class 0 scores clustered around lower values, Class 1 around higher values
y_scores[y_true == 0] = np.random.normal(loc=0.3, scale=0.1, size=sum(y_true == 0))
y_scores[y_true == 1] = np.random.normal(loc=0.7, scale=0.1, size=sum(y_true == 1))

# Add slight overlap between classes to prevent perfect AUC
y_scores[y_true == 0] = np.clip(y_scores[y_true == 0], 0.2, 0.8)
y_scores[y_true == 1] = np.clip(y_scores[y_true == 1], 0.2, 0.8)

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

# Plot formatting to match typical ROC visualization
plt.figure(figsize=(8, 6))

# Plot ROC curve
plt.plot(fpr, tpr, color='darkorange', lw=2, 
         label=f'ROC curve (AUC = {roc_auc:.2f})')

# Plot random guess line
plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--', 
         label='Random guess (AUC = 0.50)')

# Add labels and title
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=14, pad=20)

# Set axis limits and grid
plt.xlim([-0.005, 1.0])
plt.ylim([0.0, 1.005])
plt.grid(True, alpha=0.3)

# Add legend in lower right
plt.legend(loc='lower right', frameon=True)

# Add AUC text in the upper left
plt.text(0.6, 0.2, f'AUC = {roc_auc:.2f}', 
         ha='center', va='center', fontsize=12,
         bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```
